```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import datasets
```

```python
boston = datasets.load_boston()
print(boston.DESCR)
```

## Q1: Save the dataset into pandas dataframe

```python
df_boston = pd.DataFrame(boston.data, columns= boston.feature_names)
df_boston['Price'] = boston.target
df_boston.head()
```

## Q2: Use different histograms to plot features that have right, left and zero skew.

```python
sns.histplot(df_boston, x ='PTRATIO' ,kde = True) 
```

```python
sns.histplot(df_boston, x ='RM', kde = True)  
```

```python
sns.histplot(df_boston, x ='Price', bins=20, kde = True)  
```

## Q3: Prove your previous result using mean, median and mode.

```python
print('Mean:',df_boston['PTRATIO'].mean())
print('Median:',df_boston['PTRATIO'].median())
print('Mode:',df_boston['PTRATIO'].mode())
print('Mean < Median, Left skewed Distribution - Negative')
```

```python
print('Mean:',df_boston['RM'].mean())
print('Median:',df_boston['RM'].median())
print('Mode:',df_boston['RM'].mode())
print('Mean = Median = Mode, No skewed ')
```

```python
print('Mean =',df_boston['Price'].mean())
print('Median =',df_boston['Price'].median())
print('Mode =',df_boston['Price'].mode())
print('Mean > Median, Right skewed Distribution - Positive')
```

## Q8: Simulate Two Coin Flips, then draw the result distribution.

```python
def coinFlip(p):    
    result = np.random.binomial(1,p)
    return result

probability = .2
n = 10
fullResults = np.arange(n)
for i in range(0, n):    
    fullResults[i] = coinFlip(probability)    
    i+=1
print("Head Count: ", np.count_nonzero(fullResults == 1))
print("Tail Count: ", np.count_nonzero(fullResults == 0))
sns.histplot(fullResults)
```
